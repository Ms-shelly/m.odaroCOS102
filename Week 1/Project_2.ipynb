```python
# b) Compound Interest
print("Compound Interest Calculator")

 ![Compound Interest Flowchart](flowchart_2.png)
# Getting User's Input 
P = float(input("Enter Principal Value (P):" ))
R = float(input("Enter Rate Value (R):" ))
t = float(input("Enter Time Value (T):" ))
n = int(input("Enter number of times interest compounded per year (n):" ))

# Calculating Compound Interest
A = float(P*(1.0 + (R/n) ** (n*t)))
print("Compound Interest is :",A)

```
