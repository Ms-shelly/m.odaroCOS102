```python
# a) Simple Interest
print("Simple Interest Calculator")

# Getting User's Input 
P = float(input("Enter Principal Value (P):" ))
R = float(input("Enter Rate Value (R):" ))
T = float(input("Enter Time Value (T):" ))

#Calculating Simple Interest
A = float(P*(1.0 + (R/100.0)*T))   
print("Simple Interest is :",A)
```
