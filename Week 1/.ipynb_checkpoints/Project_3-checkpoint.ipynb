```python
# c) Annuity Plan
print("Annuity Plan Calculator")

# Getting User's Input
PMT = float(input("Enter Payment per period (PMT):" ))
R = float(input("Enter Annual Interest Rate (R):" ))
n = float(input("Enter number of times interest compounded per year (n):" ))
t = float(input("Enter Time Value (T):" ))

# Calculating Annuity Plan
A = float(PMT * ((1.0 + (R/n)) ** (n*t) - 1.0) / (R/n))
print("Annuity Plan is :", A)
```
